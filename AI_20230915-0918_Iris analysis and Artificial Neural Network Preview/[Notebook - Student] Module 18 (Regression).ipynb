# 회귀(Regression)

이 노트북에서 우리는 몇 가지 일반적인 회귀 알고리즘에 대하여 논의할 것입니다. 회귀 분석을 살펴봅니다.

### 회귀 분석이란 무엇입니까? 회귀 분석을 하는 이유는 무엇입니까?
회귀 분석은 종속 변수와 독립 변수 간의 관계를 조사합니다. 일반적으로 예측, 시계열 모델링 및 변수 간의 인과 관계를 찾을 때 사용됩니다.


예를 들어 다음 분기의 수확량을 예측하려는 농부가 있다고 가정해 보겠습니다. 강우량, 종자 품질, 토양 품질과 같이 작물 수확량에 영향을 미치는 변수가 많다는 것을 알 수 있습니다. 그 중에 어떤 변수가 작물 수확량에 가장 큰 영향을 미치는지 생각해볼 수 있습니다.

"토양의 질이 높을수록 농작물 수확량이 증가할 수 있습니다." "구름의 양이 많아질 수록 농작물 수확량이 증가할 수도 있습니다."

이제 모든 변수가 유효하다고 생각할 수도 있습니다! 유효한 관계를 어떠한 방법으로 증명할 수 있을까요?

회귀 분석을 사용할 수 있습니다. 이를 통해 변수 중 어떤 것이 작물 수확량에 영향을 미칠 수 있는지 수학적으로 결정할 수 있으므로 농부는 정보에 기반하여 전략적인 선택을 취할 수 있습니다. 예를 들어, 토양 품질이 작물에 가장 큰 영향을 미친다는 사실을 아는 것은 농부들이 농장에서 양질의 토양을 유지하기 위해 더 많은 자원을 투입하도록 동기를 부여할 것입니다. 회귀 분석은 가장 중요한 변수, 무시할 수 있는 변수 및 변수가 서로 상호 작용하는 방식을 파악하는데 도움이 됩니다.


### 상관 관계 vs 인과 관계
회귀 분석을 할 때 사람들이 가장 많이 법하는 실수 중 하나는 상관 관계를 찾아내고 인과 관계를 가정하는 것입니다. 예제를 살펴보겠습니다. 아래 그래프는 아이스크림을 먹는 사람들의 수가 상어 공격의 수와 상관관계가 있음을 보여줍니다. 이 상관관계를 어떻게 설명하시겠습니까? 아이스크림을 먹으면 상어 공격이 일어난다고 할까요, 아니면 상어의 공격이 더 많은 사람들이 아이스크림을 먹는 결과를 초래한다고 할까요?

![Image](resources/icecreamand-sharks.png)

사실 아이스크림 판매량과 상어의 공격은 인과관계가 없습니다. 사람들이 아이스크림을 먹는 것은 상어의 공격 횟수에 영향을 미치지 않습니다!

따라서 데이터를 처리하고 회귀 분석을 수행할 때 반드시 변수 간의 상관 관계가 인과 관계를 의미하는 것은 아니라는 점을 고려해야 합니다.

즉, 두 변수가 관계가 있다고 해서 한 변수가 다른 변수에 영향을 미친다는 의미는 아닙니다. 두 사건 사이의 관계를 결정하기 전에 검토해야 할 변수가 많이 있습니다. 상관 관계를 일으키는 다른 변수를 찾을 수 있습니다.

궁금한 사람들을 위해 해당 그래프는 https://www.tylervigen.com/spurious-correlations 에서 가져왔습니다. 해당 링크로 이동하여 흥미로운 상관 관계를 볼 수 있습니다.

# 선형 회귀(Linear Regression)

선형 회귀는 가장 일반적인 회귀 기법 중 하나입니다. 직선을 사용하여 변수 간의 관계를 나타냅니다. 직선으로 표현할 수 있는 관계는 무엇이 있을까요? 키 대 나이? 공부한 시간과 결과? 더 많이 있습니다!

### 회귀선
최적 적합선으로 알려진 회귀선은 선형 관계를 결정하는 데 사용됩니다. 우리가 비교하는 변수 간의 관계를 가장 잘 나타내는 선입니다.

아래 그림을 표시된 직선을 y = mx + c의 방정식으로 설명할 수 있습니다:
- m은 선의 기울기이고,
- c는 아래와 같이 y절편(선이 y축/수직선을 절단하는 위치)입니다.

![image.png](resources/concepts12.jpeg)

회귀 분석을 지원하기 위해 sklearn 라이브러리에는 LinearRegression()이라는 함수가 있습니다. 이 함수는 데이터를 사용하여 최적선이라고도 하는 선형 회귀선을 표시하거나 찾습니다.

선형 회귀는 최소 제곱법을 사용하여 최적의 선을 구합니다. 이는 y = mx + c의 방정식에서 m과 c를 구하는 것을 의미합니다. 이 경우 실제 값과 예측 값의 간의 차이에 대한 제곱의 최소 합을 구합니다.

최소 제곱법을 사용하여 선형 회귀를 계산하는 방법에 대해 자세히 알아보려면 이 비디오를 참조하세요. https://www.youtube.com/watch?v=JvS2triCgOY

## 챌린지 타임
이 글을 읽고 선형 회귀 대신 다항식 회귀를 사용하여 모델을 구현해 보십시오. https://towardsdatascience.com/linear-regression-on-boston-housing-dataset-f409b7e4a155

```python
# 다항식 회귀 모델 구현
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pandas as pd
import numpy as np

# 데이터 준비 (X와 y 설정)
features_available = ['Temperature (C)']
X = data_sep[features_available]
y = data_sep['Humidity']

# 다항식 특성 추가
degree = 2  # 2차 다항식을 사용하려면 degree를 2로 설정
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.3, random_state=42)

# 다항식 회귀 모델 훈련
linreg = LinearRegression()
linreg.fit(X_train, y_train)

# 테스트 데이터에 대한 예측 수행
y_pred = linreg.predict(X_test)

# 성능 지표 계산
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# 결과 출력
df1 = pd.DataFrame(columns=['Actual', 'Predicted'])
df1['Actual'] = y_test
df1['Predicted'] = y_pred

print(df1)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared (R2) Score:", r2)
```
