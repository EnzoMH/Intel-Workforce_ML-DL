```python
# 분류를 위해 특성(평균 색상) 및 해당 레이블(빨간색/녹색/검정색/없음) 저장
# 이미지 분류를 위해 특성(평균 색상)을 추출하고 해당 레이블(빨간색/녹색/검정색/없음)을 저장하는 작업을 수행
trainX = [] # trainX 리스트를 초기화합니다. 이 리스트는 이미지 특성과 레이블을 저장하기 위한 빈 리스트
trainY = [] # trainY 리스트를 초기화합니다. 이 리스트는 이미지 특성과 레이블을 저장하기 위한 빈 리스트

# 카드들에 대해 반복하고 평균 색상 출력
# for 루프를 사용하여 네 가지 종류의 카드 (빨간색, 녹색, 검정색, 배경)에 대해 반복
# zip() 함수를 사용하여 각 카드 이미지와 해당 레이블을 쌍으로 묶습니다.
for (banana,label) in zip((new_banana,unripe_banana,rotten_banana),("reapable banana","unreapable banana","rotten banana")):   

# 현재 처리 중인 카드의 레이블과 해당 카드의 평균 색상을 출력합니다. 
# averagecolor() 함수는 이미지의 평균 색상을 계산하고, 이를 (label, 평균 색상) 형식으로 출력합니다
    print((label, averagecolor(banana)))
        
#     평균 색상을 계산한 후, trainX 리스트에 평균 색상을 추가합니다. 
#     이렇게 하면 각 카드의 평균 색상이 trainX 리스트에 저장됩니다.

    trainX.append(averagecolor(banana))
    
#     현재 처리 중인 카드의 레이블을 trainY 리스트에 추가합니다. 
#     이렇게 하면 각 카드에 해당하는 레이블이 trainY 리스트에 저장됩니다.

    trainY.append(label)
```

```python
# 1. 이미지 및 함수 불러오기:

import cv2
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

# 2.새로운 카드 이미지를 읽고 평균 색상 계산:

new_banana = cv2.imread("C:/Users/isfs0/Intel_AI Process/AI_20231010/BananaClassification/ripe/Image_1.jpg")
new_banana_features = averagecolor(new_banana)
unripe_banana = cv2.imread("C:/Users\isfs0/Intel_AI Process/AI_20231010/BananaClassification/unripe/gImage_1.jpg")
unripe_banana_features = averagecolor(unripe_banana)
rotten_banana = cv2.imread("C:/Users\isfs0/Intel_AI Process/AI_20231010/BananaClassification/overripe/rImage_2.jpg")
rotten_banana_features = averagecolor(rotten_banana)

def averagecolor(image): # 함수 선언 averagecolor, and 색깔의 평균값
    return np.mean(image, axis=(0, 1))


# new_banana = cv2.imread("C:/Users/isfs0/Intel_AI Process/AI_20231010/BananaClassification/ripe/Yellow Banana/Image_1.jpg")
# unripe_banana = cv2.imread("C:/Users\isfs0/Intel_AI Process/AI_20231010/BananaClassification/unripe/Green Banana/Image_1.jpg")
# rotten_banana = cv2.imread("C:/Users\isfs0/Intel_AI Process/AI_20231010/BananaClassification/overripe/Rotten Banana/Image_2.jpeg")
```

```python
# 3. 학습 데이터 생성:
# new_banana = cv2.imread("C:/Users/isfs0/Intel_AI Process/AI_20231010/BananaClassification/ripe/Yellow Banana/Image_1.jpg")
# unripe_banana = cv2.imread("C:/Users\isfs0/Intel_AI Process/AI_20231010/BananaClassification/unripe/Green Banana/Image_1.jpg")
# rotten_banana = cv2.imread("C:/Users\isfs0/Intel_AI Process/AI_20231010/BananaClassification/overripe/Rotten Banana/Image_2.jpg")

# 학습 데이터의 특징과 레이블 생성
trainX = []
trainY = []

for (banana, label) in zip((new_banana, unripe_banana, rotten_banana), ("Ripe Banana", "Unripe Banana", "Rotten Banana")):
    trainX.append(averagecolor(banana))
    trainY.append(label)
    print((label,averagecolor(banana)))
```

```python
# 4. 테스트 이미지에 대한 거리 계산 및 예측(유클라디안):


calculated_distances = []
for banana in (trainX):
    calculated_distances.append(np.linalg.norm(new_banana_features-banana))
    
print(calculated_distances)
```

```python

from sklearn.metrics import classification_report

realtestY = np.array(["Ripe Banana","Ripe Banana","Ripe Banana","Ripe Banana",
                     "Ripe Banana","Ripe Banana","Ripe Banana","Ripe Banana","Ripe Banana","Ripe Banana","Unripe Banana",
                     "Unripe Banana","Unripe Banana","Unripe Banana","Unripe Banana","Unripe Banana",
                     "Unripe Banana","Unripe Banana","Unripe Banana","Rotten Banana","Rotten Banana","Rotten Banana",
                     "Rotten Banana","Rotten Banana","Rotten Banana","Rotten Banana","Rotten Banana","Rotten Banana",
                     "Rotten Banana","Rotten Banana","Rotten Banana","Rotten Banana","Rotten Banana","Rotten Banana"])
# 실제 테스트 이미지에 대한 정답 레이블을 나타내는 NumPy 배열을 생성합니다.
# 예를 들어, 첫 번째 이미지는 "black"으로 레이블링되었으며, 마지막 이미지는 "none"으로 레이블링되었습니다.

def evaluateaccuracy(filenames, predictedY):
    predictedY = np.array(predictedY)
    # 예측된 레이블을 NumPy 배열로 변환합니다.
    
    if (np.sum(realtestY != predictedY) > 0):
        # 만약 예측된 레이블과 실제 레이블이 다른 경우:
        print ("Wrong Predictions: (filename, labelled, predicted) ")
        # "Wrong Predictions"과 함께 파일 이름, 실제 레이블, 예측된 레이블을 출력합니다.
        print (np.dstack([filenames, realtestY, predictedY]).squeeze()[(realtestY != predictedY)])
        # 잘못 분류된 항목의 파일 이름, 실제 레이블, 예측된 레이블을 출력합니다.
    
    # 전체 예측의 백분율로 일치하는 (정확한) 예측을 계산합니다.
    return "Correct :" + str(np.sum(realtestY == predictedY)) + ". Wrong: " + str(np.sum(realtestY != predictedY)) + ". Correctly Classified: " + str(np.sum(realtestY == predictedY) * 100 / len(predictedY)) + "%"
```

```python
# os 라이브러리를 사용하여 테스트 이미지가 있는 폴더 경로를 지정하고, 
# 빈 리스트 predictedY와 filenames를 초기화합니다.

import os
path = "C:/Users/isfs0/Intel_AI Process/AI_20231010/BananaClassification/Test/"
predictedY = []
filenames = []

# 지정된 폴더 내의 파일 목록을 순환하며 파일 이름을 filename 변수에 할당합니다.
for filename in os.listdir(path):
    
    img = cv2.imread(path+filename)
    
    img_features = averagecolor(img)
    
    calculated_distances = []
    
    for banana in (trainX):
        calculated_distances.append(np.linalg.norm(img_features-banana))
        prediction = trainY[np.argmin(calculated_distances)]
    
    print (filename + ": " + prediction) # 추론을 출력합니다.
    filenames.append(filename)
    predictedY.append(prediction)


print ()
print(classification_report(realtestY, predictedY))


print ()
print (evaluateaccuracy(filenames,predictedY))
```
