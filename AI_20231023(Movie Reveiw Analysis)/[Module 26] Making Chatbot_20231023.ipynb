```python
# 3-3) 표제어 추출함수 정의
lemmer = nltk.stem.WordNetLemmatizer() # lemmer 클래스를 초기화합니다. WordNet은 NLTK에 포함된 의미 론적 영어 사전입니다
def LemTokens(tokens):
    return [lemmer.lemmatize(token) for token in tokens]
```

```python
# 3-4) 정규화 함수 정의
remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)
def LemNormalize(text):
    return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict))) #see previous section 1.2.5 lemmatization
```

# 4. 응답함수 정의하기

```python
# 전체코드
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 4. 응답함수 정의하기

def response(user_input):
    # 4.1) 질문 문장을 기존 토큰에 추가하기
    sentences.append(user_input)

    # 4.2) 토큰에 대해 tfidf 생성하기
    tfidf_vectorizer = TfidfVectorizer(tokenizer=LemNormalize, stop_words='english')
    tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)

    # 4.3) 질문 문장과 기존 토큰과의 코사인 유사도 구하기
    cosine_sim = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])

    # 4.4) 코사인 유사도 값이 가장 큰 값 구하기
    max_cosine_sim = np.argmax(cosine_sim)

    # 4.5) 코사인 유사도 값이 가장 큰 문장 + 유사도값 출력하기
    response = sentences[max_cosine_sim]

    return response

# # 테스트
# user_question = "What's the weather like today?"
# bot_response = response(user_question)
# print("챗봇 응답:", bot_response)
```

# 5. 인삿말 대응함수 정의하기 - greeting()

```python
GREETING_INPUTS = ["안녕", "뭐해", "머해", "야", "?","대답해봐", " "]
GREETING_RESPONSES = ["안녕하세요!", "반갑습니다", "반가워요!", "안녕하냐, 난 챗봇이다", "뭔데 넌", "반갑다. 인간시대의 끝이 도래했따."]
```

```python
# 대응함수 개인화 
```

```python
ADDITIONAL_GREETING_INPUTS = ["뭐야이건", "뭐야 이건", "반갑다", "반가워", "?","??", " "]
ADDITIONAL_GREETING_RESPONSES = ["뭐긴 뭐야 챗봇이지", "챗봇입니다.", "반가워요!", "반갑습니다", "?"]

# 기존 목록에 추가
GREETING_INPUTS.extend(ADDITIONAL_GREETING_INPUTS)
GREETING_RESPONSES.extend(ADDITIONAL_GREETING_RESPONSES)
```

```python
def greeting(sentence):
    for word in sentence.split(): # 문장의 각 단어를 살펴봅니다.
        if word.lower() in GREETING_INPUTS: # 단어가 GREETING_INPUT와 일치하는지 확인합니다.
            return random.choice(GREETING_RESPONSES) # Greeting_Response로 답장합니다.
```

```python
def response(user_response):
    
    robo_response='' # 문자열을 포함하도록 변수를 초기화
    sent_tokens.append(user_response) # sent_messages에 사용자 응답 추가
    TfidfVec = TfidfVectorizer(tokenizer=LemNormalize, stop_words='english') 
    tfidf = TfidfVec.fit_transform(sent_tokens) # tfidf 값 가져오기

    vals = cosine_similarity(tfidf[-1], tfidf) # 코사인 유사성 값 가져오기
    idx=vals.argsort()[0][-2] 
    
    flat = vals.flatten() 
    flat.sort() # 오름차순으로 정렬
    req_tfidf = flat[-2] 
    
    if(req_tfidf==0):
        robo_response=robo_response+"I am sorry! I don't understand you"
        return robo_response
    
    else:
        robo_response = robo_response+sent_tokens[idx]
        return robo_response
```

# 6. 끝인삿말 대응 함수 정의하기 - goodbye()

```python
# 끝인삿말 대응 함수 정의
def goodbye(sentence):
    if 'bye' in sentence.lower():
        return "Goodbye! Have a nice day."
    return None

def classify_question(user_input):
    # 입력 문자열을 소문자로 변환
    user_input = user_input.lower()

    # '점심' 또는 '저녁'과 '먹'이 포함된 경우
    if ('lunch' in user_input or 'dinner' in user_input) and 'eat' or 'ate' in user_input:
        return "I can't eat because I'm a machine."
    
    # '날씨'와 '궁금해', '알려줘', '어때'가 포함된 경우
    if 'weather' in user_input and ('imformation' in user_input or "what's the" in user_input or "how is it" in user_input):
        return "Today weather is warm and cool, Today will have 21 celcious degree and 11 celcisous degree."
    
    # 다른 질문 유형을 여기에 추가할 수 있습니다.
    
    # 기본적으로 일반적인 응답을 반환
    return "what food did you eat?"


# 테스트를 위한 챗봇 사용 예시
def chat_with_bot():
    print("Chatbot : Hello, how can I help you? (종료하려면 'bye'를 입력하세요)")

    while True:
        user_input = input(": ")
        user_input = user_input.lower()

        # 인삿말 대응
        if greeting(user_input) is not None:
            print("Chatbot: " + greeting(user_input))
            continue

        # 끝인삿말 대응
        if goodbye(user_input) is not None:
            print("Chatbot: " + goodbye(user_input))
            break

        # 사용자 질문에 대한 응답 생성 및 출력
        bot_response = response(user_input)
        print("Chatbot: " + bot_response)

# 챗봇 사용 예시 실행
chat_with_bot()
```

# 7. 챗봇 사용하기

```python
# 7-1) 안내문 띄우기
def introduction():
    print("챗봇: 안녕! 무엇을 도와드릴까요? (종료하려면 'bye'를 입력하세요)")
```

```python
# 7-2) 질문 입력받기
def get_user_input():
    return input("사용자: ")
```

```python
# 7-3) 질문 소문자 변경하기
def preprocess_input(user_input):
    return user_input.lower()
```

```python
# 7- 4) 인삿말 - 인사말대응, 끝인삿말-끝인삿말대응, 질문-챗봇응대
def chat_with_bot():
    introduction()  # 안내문 띄우기

    while True:
        user_input = get_user_input()  # 질문 입력받기
        user_input = preprocess_input(user_input)  # 질문 소문자 변경

        # 인삿말 대응
        if greeting(user_input) is not None:
            print("챗봇: " + greeting(user_input))
            continue

        # 끝인삿말 대응
        if goodbye(user_input) is not None:
            print("챗봇: " + goodbye(user_input))
            break

        # 사용자 질문에 대한 응답 생성 및 출력
        bot_response = response(user_input)
        print("챗봇: " + bot_response)
```

```python
# 7. 챗봇 사용하기
chat_with_bot()
```
